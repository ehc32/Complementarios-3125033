```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import pearsonr

# Datos
data = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Número de desarrolladores (millones)': [23.00, 23.90, 24.50, 25.60, 26.40, 27.70, 28.70]
}

# Cargar datos en un DataFrame
df = pd.DataFrame(data)

# Separar variables independientes (X) y dependientes (y)
X = df[['Año']].values
y = df['Número de desarrolladores (millones)'].values

# Calcular correlación de Pearson
correlation, p_value = pearsonr(df['Año'], df['Número de desarrolladores (millones)'])

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predicciones
y_pred = model.predict(X)

# Evaluar el modelo
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Imprimir resultados
print("Coeficiente de la regresión (pendiente):", model.coef_[0])
print("Intercepto:", model.intercept_)
print("Error cuadrático medio (MSE):", mse)
print("Coeficiente de determinación (R^2):", r2)
print("Correlación de Pearson:", correlation)
print("Valor p de la correlación:", p_value)

# Visualizar resultados
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Línea de regresión')
plt.title('Regresión Lineal: Número de desarrolladores por año')
plt.xlabel('Año')
plt.ylabel('Número de desarrolladores (millones)')
plt.legend()
plt.grid()
plt.show()


# Guardar la gráfica como un archivo
plt.savefig("grafica_regresion.png")

```
